```python
from framsfiles import reader as framsreader
from emd_on_voxels import VoxelsEMD
import pandas as pd
import seaborn as sns
import matplotlib.p
```

```python
genos = framsreader.load("/home/adam/Framsticks/Framsticks50rc19/data/walking.gen", "gen file")
```

```python
path = '/home/adam/Framsticks/Framsticks50rc19'
e = VoxelsEMD(path)
```

```python
print("Calculating EMDforGeno")
# list_of_genos = ["X","XX","XXX","XXXXX","XXXXCCX","XCCXXXCCX"]
list_of_genos = [x['genotype'] for x in genos]
dissim = e.getDissimilarityMatrix(list_of_genos)
```

```python
def drawModel(modelVoxels):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')

    xs = modelVoxels[:,0]
    mx = max(xs)
    ys = modelVoxels[:,1]
    my = max(ys)
    zs = modelVoxels[:,2]
    mz = max(zs)
    max_v = max([mx,my,mz])
    ax.scatter(xs, ys, zs)
    ax.set_xlim([-max_v, max_v])
    ax.set_ylim([-max_v, max_v])
    ax.set_zlim([-max_v, max_v])
    print(plt.isinteractive())
    plt.show()

def drawModel2(modelVoxels,modelVoxels2):
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(121, projection='3d')

    xs = modelVoxels[:,0]
    ys = modelVoxels[:,1]
    zs = modelVoxels[:,2]

    mx = max(xs)
    my = max(ys)
    mz = max(zs)
    max1_v = max([mx,my,mz])
    
    ax.scatter(xs, ys, zs)

    ax2 = fig.add_subplot(122, projection='3d')

    xs = modelVoxels2[:,0]
    ys = modelVoxels2[:,1]
    zs = modelVoxels2[:,2]
    
    mx = max(xs)
    my = max(ys)
    mz = max(zs)
    
    max2_v = max([mx,my,mz])
    max_both = max([max1_v,max2_v])
    ax2.scatter(xs, ys, zs)
    
    ax.set_xlim([-max_both, max_both])
    ax.set_ylim([-max_both, max_both])
    ax.set_zlim([-max_both, max_both])

    ax2.set_xlim([-max_both, max_both])
    ax2.set_ylim([-max_both, max_both])
    ax2.set_zlim([-max_both, max_both])

    plt.show()

    
def printModel(genotype):
    voxels = e.getVoxels(genotype)
    drawModel(voxels)
    
def printModels(genotype1, genotype2):
    voxels1 = e.getVoxels(genotype1)
    voxels2 = e.getVoxels(genotype2)
    drawModel2(voxels1,voxels2)
    
# printModel(list_of_genos[0])
```
