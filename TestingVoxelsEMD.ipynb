```python
from framsfiles import reader as framsreader
from emd_on_voxels import VoxelsEMD
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utils import *
```

```python
path = '/home/adam/Framsticks/Framsticks50rc19'
e = VoxelsEMD(path)
```

```python
def printModel(genotype,name=None):
    voxels = e.getVoxels(genotype)
    if name != None:
        drawModel(voxels,name)
    else:
        drawModel(voxels)
    
def printModels(genotype1, genotype2,name=None):
    voxels1 = e.getVoxels(genotype1)
    voxels2 = e.getVoxels(genotype2)
    if name!=None:
        drawModel2(voxels1,voxels2,name)
    else:
        drawModel2(voxels1,voxels2)
    
```

```python
walking_gen = framsreader.load(path + "/data/walking.gen", "gen file")
f4_best_gen = framsreader.load(path + "/data/encoding_f4_best.gen", "gen file")
```

```python
list_of_genos_walking = [x['genotype'] for x in walking_gen]
dissim_walking = e.getDissimilarityMatrix(list_of_genos_walking)
```

# Visualisation of dissimilarity matrix using heatmap

```python
fig = plt.figure(figsize=(18,14))
sns.heatmap(dissim_walking)
plt.savefig("images/dissim_walking_heatmap.pdf")
plt.show()
```

```python
list_of_f4_best_gen = [x['genotype'] for x in f4_best_gen]
list_of_f4_best_gen.pop(14)
dissim_f4_best_gen = e.getDissimilarityMatrix(list_of_f4_best_gen)
```

```python
fig = plt.figure(figsize=(18,14))
sns.heatmap(dissim_f4_best_gen)
plt.savefig("images/dissim_f4_best_gen_heatmap.pdf")
plt.show()
```
